```python
import numpy as np
import urllib, json

import json
```

```python
nameslista = []
ralista = []
declista = []
with open('listamcvn.txt','r') as file:
    for line in file:
        l = line.split(',')
        
        decl = float(l[2].strip())
        #if decl < 0.0:
        nameslista.append(l[0])
        ralista.append(float(l[1]))
        declista.append(decl)
            

ralista = np.array(ralista)   
```

```python
urllist = []
for r,d in zip(ralista,declista):
    url = "https://mast.stsci.edu/tesscut/api/v0.1/sector?ra={ra}&dec={dec}&y=15&x=10&units=px&sector=All".format(ra=r,dec=d)
    urllist.append(url)
    print(url)
```

```python
responsedic = {}
for name,url,rad,decd in zip(nameslista,urllist,ralista,declista):
    response = urllib.request.urlopen(url)
    data = json.loads(response.read())
    print(name,rad,decd,data,len(data['results']))
    subdic = {}
    subdic['ra'] = rad
    subdic['dec'] = decd
    subdic['results'] = data
    subdic['resultsleng'] = len(data['results'])
    if len(data['results']) > 0:
        sectors = []
        for i in data['results']:
            sectors.append(int(i['sector']))
            subdic['sectors'] = sectors
    else:
        sectors = []
    print(sectors)
   
    responsedic[name] = subdic
   
```

# Save Python dic to json

```python
jsonval = json.dumps(responsedic)
f = open("response.json","w")
f.write(jsonval)
f.close()
```

```python
with open('text.txt','w') as file:
    for i in responsedic:
        resultsleng = responsedic[i]['resultsleng']
        if resultsleng > 0:
            text = '{} {}'.format(i,responsedic[i]['sectors'])
        else:
            text = '{} {}'.format(i,resultsleng)
        print(text)
        file.write('{}\n'.format(text))
    
```
